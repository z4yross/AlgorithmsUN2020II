#RECURSIVE FIBONACCI

```
import matplotlib.pyplot as plt
from timeit import default_timer as timer


def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

times = []
xs = []

for x in range(0, 50 + 1):
  start = timer()
  fibonacci(x)
  end = timer()
  xs.append(x)
  times.append(end - start)

plt.title('Recursive fibonacci(n) vs time')
plt.grid()
plt.plot(xs, times)
```

#ITERATIVE FIBONACCI

```
def fib(n):
    a = 0
    b = 1
    
    for k in range(n):
        c = b + a
        a = b
        b = c
        
    return a

times = []
xs = []

for x in range(0, 500 + 1):
  start = timer()
  fib(x)
  end = timer()
  xs.append(x)
  times.append(end - start)

plt.title('Iterative fibonacci(n) vs time')
plt.grid()
plt.plot(xs, times)
```
